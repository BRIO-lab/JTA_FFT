```python
import numpy as np
import os
import glob
from JTA_FFT import JTA_FFT
import cv2
```

```python
HOME_DIR  = "C:/Datasets_FemCleaned/Lima/Lima_Organized/Patient 77-06-HL/Session_1/"
dirlist = [filename for filename in os.listdir(HOME_DIR) if os.path.isdir(os.path.join(HOME_DIR,filename))]
lib_dc = np.load(HOME_DIR + "/dc_tib.npy")
lib_angle = np.load(HOME_DIR + "/lib-angle_tib.npy")
lib_mag = np.load(HOME_DIR + "/mag_tib.npy")
lib_surface = np.load(HOME_DIR + "/surface_tib.npy")
fem_surface = np.load(HOME_DIR + "surface_tib.npy")
rot_indices = np.load(HOME_DIR + "/rot_indices.npy")

for mvt_id in dirlist:
    MVT_DIR = HOME_DIR + "/" + mvt_id 
    IMG_DIR = MVT_DIR + "/1024/tib/" 
    num = len(os.listdir(IMG_DIR))
    jts_file = np.empty([num,6])
    for idx, img_id in enumerate(os.listdir(IMG_DIR)):
        img = cv2.imread(IMG_DIR + "/" + img_id,cv2.IMREAD_GRAYSCALE)
        cal = MVT_DIR + "/calibration.txt"
        TIB_FFT = JTA_FFT(cal)
        
        x_input, y_input = TIB_FFT.create_contour(img)
        dc, mag, angle, surface = TIB_FFT.get_NFD(x_input, y_input)

        xt, yt, zt, zr, xr, yr = TIB_FFT.estimate_pose(rot_indices, lib_dc, lib_mag, lib_angle, lib_surface, dc, mag, angle, surface)

        jts_file[idx,:] = [xt,yt,zt,zr,xr,yr]

    
    header = ["x_tran", "y_tran", "z_tran", "z_rot", "x_rot", "y_rot"]

    np.savetxt(MVT_DIR + "/tib_fft.jts", jts_file, delimiter=',', header = "x_trans, y_trans, z_trans, z_rot, x_rot, y_rot", fmt = "   %.2f")
    
```
